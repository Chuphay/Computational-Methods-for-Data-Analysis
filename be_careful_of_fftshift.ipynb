```python
import matplotlib.pylab as pylab
pylab.rcParams['figure.figsize'] = 16, 12
```

# Loading Image Data:

```python
from PIL import Image
import matplotlib.cm as cm
im = Image.open("coffee.jpg").convert('L')
im_array = pylab.array(im)
print im_array.shape
pylab.imshow(im_array,cmap = cm.Greys_r)
```

# FFT & inversed FFT:

```python
import numpy as np
F = np.fft.fft2(im)
ff = F.copy() 
im_new = np.fft.ifft2(ff).real
pylab.imshow(im_new,cmap=cm.Greys_r)
```

# Compute $l_\infty$ - norm error:

```python
np.max(np.abs(im_new - im))
```

# Twice fftshift along 1-axes:

```python
im_new = np.fft.ifft2(np.fft.fftshift(np.fft.fftshift(ff,axes=(1,)),axes=(1,))).real
pylab.imshow(im_new,cmap=cm.Greys_r)
print "np.max(np.abs(im_new - im)) = ",np.max(np.abs(im_new - im))
```

# Twice fftshift along 0-axes:

```python
im_new = np.fft.ifft2(np.fft.fftshift(np.fft.fftshift(ff,axes=(0,)),axes=(0,))).real
pylab.imshow(im_new,cmap=cm.Greys_r)
print "np.max(np.abs(im_new - im)) = ",np.max(np.abs(im_new - im))
```

# fftshift on odd-dim and even-dim are much different !!!
