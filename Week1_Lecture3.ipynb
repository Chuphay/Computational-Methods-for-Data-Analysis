# Video: W01_L03_P01

```python
import numpy as np
import pylab as pl
```

```python
T = 30
n = 512

t2 = np.linspace(-T/2,T/2,n+1)
t = t2[:-1]
k = (2*np.pi/T)*np.fft.fftfreq(n,1.0/n)
ks = np.fft.fftshift(k)

u = 1/np.cosh(t)
ut = np.fft.fft(u)

noise = 20
utn = ut + noise*(np.random.randn(n) + 1j*np.random.randn(n))
un = np.fft.fftshift(np.fft.ifft(utn))
```

```python
pl.subplot(2,1,1)
pl.plot(t,u,"r",
        t,np.abs(un),"k")
pl.subplot(2,1,2)
pl.plot(ks,np.abs(np.fft.fftshift(ut)),"r",
        ks,np.abs(np.fft.fftshift(utn)),"k")
```

# Video: W01_L03_P02

```python
T = 30
n = 512

t2 = np.linspace(-T/2,T/2,n+1)
t = t2[:-1]
k = (2*np.pi/T)*np.fft.fftfreq(n,1.0/n)
ks = np.fft.fftshift(k)

u = 1/np.cosh(t)
ut = np.fft.fft(u)

noise = 30

ave = np.zeros(utn.shape)
realizations = 30
for i in range(realizations):
    utn = ut + noise*(np.random.randn(n) + 1j*np.random.randn(n))
    ave = ave + utn
    
ave = np.abs(np.fft.fftshift(ave/realizations))
```

```python
pl.plot(ks,np.abs(np.fft.fftshift(ut)),"r",
        ks,ave,"k")
```

```python
pl.plot(ks,np.abs(np.fft.fftshift(ut)),"r",
        ks,ave,"k",
        ks,np.abs(np.fft.fftshift(utn)),"c")
```

```python
draw_index = [1,10,20,30]
draw_color = ["y","b","g","c"]

ave = np.zeros(utn.shape)
realizations = 30
for i in range(realizations):
    utn = ut + noise*(np.random.randn(n) + 1j*np.random.randn(n))
    ave = ave + utn
    
    if i+1 in draw_index:
        ave2 = np.abs(np.fft.fftshift(ave/(i+1)))
        pl.plot(ks,np.abs(np.fft.fftshift(ut)),"r",
                ks,ave2,draw_color[draw_index.index(i+1)])
    
#     pl.pause(0.5)
    
```

# Video: W01_L03_P03

```python
T = 60
n = 512

t2 = np.linspace(-T/2,T/2,n+1)
t = t2[:-1]
k = (2*np.pi/T)*np.fft.fftfreq(n,1.0/n)
ks = np.fft.fftshift(k)
```

```python
_slice = np.arange(0,10,0.5)
_slice
```

```python
TT,TS = np.meshgrid(t,_slice)
TT,TS
```

```python
K,KS = np.meshgrid(k,_slice)
K,KS
```

```python
U = 1/np.cosh(TT - 10*np.sin(TS))*np.exp(1j*0*TT)
U.shape
```

```python
from mpl_toolkits.mplot3d import axes3d
import matplotlib.pyplot as plt
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_wireframe(TT, TS, U, rstride=10, cstride=10)

plt.show()
```

```python
UT = np.zeros(U.shape)
```

```python
for ii in range(len(_slice)):
    UT[ii,:] = np.abs(np.fft.fftshift(np.fft.fft(U[ii,:])))

fig = plt.figure()
ax = fig.add_subplot(211, projection='3d')
ax.plot_wireframe(TT, TS, U, rstride=10, cstride=10)
ax = fig.add_subplot(212, projection='3d')
ax.plot_wireframe(TT, TS, UT, rstride=10, cstride=10)

plt.show()
                
```
