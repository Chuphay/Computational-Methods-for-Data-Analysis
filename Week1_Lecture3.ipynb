# Video: W01_L03_P01

```python
import numpy as np
import pylab as pl
```

```python
T = 30
n = 512

t2 = np.linspace(-T/2,T/2,n+1)
t = t2[:-1]
k = (2*np.pi/T)*np.fft.fftfreq(n,1.0/n)
ks = np.fft.fftshift(k)

u = 1/np.cosh(t)
ut = np.fft.fft(u)

noise = 20
utn = ut + noise*(np.random.randn(n) + 1j*np.random.randn(n))
un = np.fft.fftshift(np.fft.ifft(utn))
```

```python
pl.subplot(2,1,1)
pl.plot(t,u,"r",
        t,np.abs(un),"k")
pl.subplot(2,1,2)
pl.plot(ks,np.abs(np.fft.fftshift(ut)),"r",
        ks,np.abs(np.fft.fftshift(utn)),"k")
```

# Video: W01_L03_P04

```python
T = 60
n = 512

t2 = np.linspace(-T/2,T/2,n+1)
t = t2[:-1]
k = (2*np.pi/T)*np.fft.fftfreq(n,1.0/n)
ks = np.fft.fftshift(k)

_slice = np.arange(0,10,0.5)

TT,TS = np.meshgrid(t,_slice)
K,KS = np.meshgrid(k,_slice)

U = 1/np.cosh(TT - 10*np.sin(TS))*np.exp(1j*10*TT)

UT = np.zeros(U.shape)

for ii in range(len(_slice)):
    UT[ii,:] = np.abs(np.fft.fftshift(np.fft.fft(U[ii,:])))

fig = plt.figure()


ax = fig.add_subplot(211, projection='3d')
ax.plot_wireframe(TT, TS, U, rstride=10, cstride=10)
ax.view_init(50,-120)
ax = fig.add_subplot(212, projection='3d')
ax.plot_wireframe(np.fft.fftshift(K), KS, UT, rstride=10, cstride=10)
ax.view_init(50,-120)

plt.show()
```

```python
T = 60
n = 512

t2 = np.linspace(-T/2,T/2,n+1)
t = t2[:-1]
k = (2*np.pi/T)*np.fft.fftfreq(n,1.0/n)
ks = np.fft.fftshift(k)

_slice = np.arange(0,10,0.5)

TT,TS = np.meshgrid(t,_slice)
K,KS = np.meshgrid(k,_slice)

U = 1/np.cosh(TT - 10*np.sin(TS))*np.exp(1j*10*TT)

UT = np.zeros(U.shape)

for ii in range(len(_slice)):
    UT[ii,:] = np.abs(np.fft.fftshift(np.fft.fft(U[ii,:])))

fig = plt.figure()


ax = fig.add_subplot(211, projection='3d')
ax.plot_wireframe(TT, TS, np.abs(U), rstride=10, cstride=10)
ax.view_init(50,-120)
ax = fig.add_subplot(212, projection='3d')
ax.plot_wireframe(np.fft.fftshift(K), KS, UT, rstride=10, cstride=10)
ax.view_init(50,-120)

plt.show()
```

## Add noise to frequency domain ...

```python
T = 60
n = 512

t2 = np.linspace(-T/2,T/2,n+1)
t = t2[:-1]
k = (2*np.pi/T)*np.fft.fftfreq(n,1.0/n)
ks = np.fft.fftshift(k)

_slice = np.arange(0,10,0.5)

TT,TS = np.meshgrid(t,_slice)
K,KS = np.meshgrid(k,_slice)

U = 1/np.cosh(TT - 10*np.sin(TS))*np.exp(1j*10*TT)

UT = np.zeros(U.shape)

noise=20
for ii in range(len(_slice)):
    UT[ii,:] = np.abs(np.fft.fftshift(np.fft.fft(U[ii,:])+noise*(np.random.randn(n) + 1j*np.random.randn(n))))

fig = plt.figure()


ax = fig.add_subplot(211, projection='3d')
ax.plot_wireframe(TT, TS, np.abs(U), rstride=10, cstride=10)
ax.view_init(50,-120)
ax = fig.add_subplot(212, projection='3d')
ax.plot_wireframe(np.fft.fftshift(K), KS, UT, rstride=10, cstride=10)
ax.view_init(50,-120)

plt.show()
```

## Add noise to frequency domain ... and ifft to time domain ...

```python
T = 60
n = 512

t2 = np.linspace(-T/2,T/2,n+1)
t = t2[:-1]
k = (2*np.pi/T)*np.fft.fftfreq(n,1.0/n)
ks = np.fft.fftshift(k)

_slice = np.arange(0,10,0.5)

TT,TS = np.meshgrid(t,_slice)
K,KS = np.meshgrid(k,_slice)

U = 1/np.cosh(TT - 10*np.sin(TS))*np.exp(1j*10*TT)

UT = np.zeros(U.shape)
UN = np.zeros(U.shape)

noise=20
for ii in range(len(_slice)):
    UT[ii,:] = np.fft.fft(U[ii,:])+noise*(np.random.randn(n) + 1j*np.random.randn(n))
    UN[ii,:] = np.fft.ifft(UT[ii,:])
fig = plt.figure()


ax = fig.add_subplot(211, projection='3d')
ax.plot_wireframe(TT, TS, np.abs(UN), rstride=10, cstride=10)
ax.view_init(50,-120)
ax = fig.add_subplot(212, projection='3d')
ax.plot_wireframe(np.fft.fftshift(K), KS, np.abs(np.fft.fftshift(UT)), rstride=10, cstride=10)
ax.view_init(50,-120)

plt.show()
```
