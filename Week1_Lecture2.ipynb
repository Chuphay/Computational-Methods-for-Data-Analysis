# Video: W01_L02_P02

```python
import numpy as np
import pylab as pl
```

```python
T = 30
n = 512
```

```python
t2 = np.linspace(-T/2,T/2,n+1)
t = t2[:-1]
k = (2*np.pi/T)*np.fft.fftfreq(n,1.0/n)
```

```python
u = 1/np.cosh(t)
pl.plot(t,u)
```

```python
ks = np.fft.fftshift(k)
ut = np.fft.fft(u)
pl.subplot(2,1,1)
pl.plot(t,u)
pl.subplot(2,1,2)
pl.plot(ks,np.abs(np.fft.fftshift(ut)))
```

# Fourier Transform Pairs:

- ## Gaussian (time) <=> Gaussian (freq)
- ## Heisenberg Uncertainty principle
- ## width of pulse in time <=> width of pulse in freq


```python
pl.subplot(2,1,1)
pl.plot(t,u)
pl.subplot(2,1,2)
pl.plot(ks,np.abs(np.fft.fftshift(ut))/max(np.abs(np.fft.fftshift(ut))))
pl.xlim([-25,25])
pl.ylim([0,1])
```
